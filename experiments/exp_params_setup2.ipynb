# Doc2Vecのhyper parameter tuning（setup2）

## 目的
setup1とは異なるセットアップ（前処理2で処理したベクトルデータ）で、Doc2Vecをトレーニング。  
(setup1の結果とをフェアに比較したいので), ここではsetup2のhyper parameter tuningを行う。

## データ
前処理2を施したlivedoor news data  
（"check_preproc2.ipynb"）

## 手段
"exp_params_setup1.ipynb"の結果より、変更して意味がありそうなhyper parameterのみをtuningする。  
具体的には、以下のパラメータをtuningする。
- epoch
- sample
- min_count
- vector_size

## 結果
tuningした結果、setup1で最適化したパラメータと同じで、ほぼ最適化の結果が得らることがわかった。

## 結論
**前処理のやり方を変更しても、最適化したパラメータセットはほぼそのまま使える。**　  
 （少くとも、このデータセットに関しては）

### Default

```python
import os, sys
import time
sys.path.append(os.path.pardir)
from utils import evaluation_docs, experiments
```

```python
%%time
tag_pp = 'pp2'
tag_setup = '2-default'
save_path='results/%s' % tag_setup

ep = experiments.ExpPar(tag_pp, tag_setup)
model = ep.train_model()
ed = evaluation_docs.evaluation_labeled_docs(model, ep.tagged_words_list, n_clusters=20, 
                                                                                  dic_category_map=ep.dic_category_map)
score, history = ed.eval_sv(save_path=save_path)
```

### a. epoch

| epoch　|　10　|　30　|　50　| 100 |
|:----|-----:|-----:|-----:|-----:|----:|
| accuracy | 0.82 | 0.86 | 0.87 | 0.86 |

```python
tag_pp = 'pp2'
tag_setups = ['2-a%d' % i for i in range(1, 5)]

for tag_setup in tag_setups:
    t0 = time.time()
    print('\n--- %s ---' % tag_setup)
    save_path='results/%s' % tag_setup

    ep = experiments.ExpPar(tag_pp, tag_setup)
    model = ep.train_model()
    ed = evaluation_docs.evaluation_labeled_docs(model, ep.tagged_words_list, n_clusters=20, 
                                                                                      dic_category_map=ep.dic_category_map)
    score, history = ed.eval_sv(save_path=save_path)
    print('Wall time: %.1f sec' % (time.time() - t0))
```
