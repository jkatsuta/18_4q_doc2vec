# 定量的な評価方法

## 目的
カテゴリデータ（= 教師データ）を用いて評価を行う  
("正解"がわかっているので、定量的な評価が可能になる)。

## モデル
doc2vec_pp1_setup1-default.model  
(train_doc2vec_1-default.ipynbでトレーニングしたモデル)

## 評価方法
#### 評価方法1
**Kmeansで学習したラベルとカテゴリの一致度**
- doc2vecが適切に学習できていれば、Kmeansのラベルとカテゴリは一致するはず。
- この評価方法は定量性を重視するというより、plotしつつ感覚をつかむ定性的な側面の意味合いが強い。

#### 評価方法2
**教師あり学習（MLP）の精度**
- doc_vecをfeature、カテゴリを教師データとして学習する。
- 教師あり学習の正解率（accuracy）を評価基準にする。
    - doc2vecが適切に学習できていれば、正解率が高くなるはず。


## 結果
### 評価方法1
dokujo-tsushin, movie-enter, smaxなどは、比較的よくラベル分けとカテゴリが一致している（70~80%）。   
一方で、kaden-cghannel, livedoor-hommeなどはあまり一致していない。

比較的良い結果のカテゴリは、他のカテゴリとの違いがわかりやすい？　　   
たとえば映画系なら"映画", "公開"という単語は入ってそうだし、  
ガジェット系もスペック表示などがあるので、わかりやすいのかも。

PCAとt-SNEでもplotしてみた。  
t-SNEは、上記の結果とconsistentなプロットになっているのがわかる。

### 評価方法2
#### 2-1. 正解率による評価
正解率（accuracy）は79%  
悪くはない精度だが、間違いもまだまだあるという値。これは教師なしの評価の感覚ともconsistent。  
この評価方法は、doc2vecのモデル精度（= vector情報）を適切に評価できている、と言ってよさそう。

#### 2-2. 9つのカテゴリ毎にも評価値を算出
「smax, movie-enterは高く、livedoor-hommeは低い」など、  
評価方法1とconsistentな結果が得られている。

#### 2-3. 正しく予測した記事と、間違って予測した記事の比較
正解しているケースでは、「そのカテゴリらしい」記事内容を適切に予測できている。  
ある程度の精度で予測できていることがわかる。  
→doc2vecがある程度うまく学習できている。

間違いのケースは２種類
1. doc2vecが適切に学習できていない。
2. 記事内容から、カテゴリ分けするのが難しいケース（教師カテゴリと記事内容が適切にマッチしていない, etc）


## 結論
**評価方法１**は、粗い定量評価なので、適切にモデルを評価できているとは言い難い。  
この評価方法は定量的というよりは、plotしつつ感覚をつかむ定性的な側面の意味合いが強い、と言える。

**評価方法２**は、適切に定量的評価ができている。  
注意点としては、「教師カテゴリと記事内容が適切にマッチしていない」ケースもあるので、  
**doc2vecの評価を目的とする場合、正解率100%を目指すことは適切ではない。**    
むしろ100%近くなったらoverfit（記事内容ではなく、特定の単語やURLで"正解"を学習している）  
を疑った方がよさげ。  

**Parameter tuningするとき正解率を相対的に比較する**、などが適切な使い方といえる。  
また数値だけでなく、実際に結果を表示するなどして、定性的に理解する事も大事。

```python
import sys
import os.path as osp
sys.path.append(osp.pardir)
import pickle
from utils import train_docs
from utils import evaluation_docs
from gensim.models.doc2vec import Doc2Vec

fn_model = './model/doc2vec_pp1_setup1-default.model'
fn_tagged_words_list = './preproc_data/words_list_mecab_pp1.pickle'
fn_category_map = './preproc_data/category_map_pp1.pickle'
# fn_model = './model/sample_doc2vec_pp1_setup1-1.model'
# fn_tagged_words_list = './preproc_data/sample_words_list_mecab_pp1.pickle'
# fn_category_map = './preproc_data/sample_category_map_pp1.pickle'
n_clusters = 20
n_each_show = 5

model = Doc2Vec.load(fn_model)
tagged_words_list = pickle.load(open(fn_tagged_words_list, 'rb'))
dic_category_map = pickle.load(open(fn_category_map, 'rb'))

ed = evaluation_docs.evaluation_labeled_docs(model, tagged_words_list, n_clusters, dic_category_map)
```

## 1. Kmeansで学習したラベルとカテゴリの一致度

ここでは以下のように粗く評価する。
1. Kmeansで学習してラベル分けする（ここではn_cluster=20).
2. 各ラベルがどのカテゴリに最も近いかを決める。
    - 各ラベルに属する記事のカテゴリをカウントする。
    - 最もカウントが多いカテゴリをそのラベルの"カテゴリ"と見なす
    - たとえば、label=3に属する記事のカテゴリ分布が、movie-enter=31, smax=124, ...etcとなったとき、
        - 最もカウントの高いsmaxをlabel=3のカテゴリと見なす。
3. 2で分けたカテゴリごとに、accuracyを求める。
    - たとえば、smaxに属するラベルがlabel=3, 5, 10だった場合、
        - acc_smax = n_smax_category / n_all_doc (@ label=3, 5, 10)のように計算する。

### 結果
dokujo-tsushin, movie-enter, smaxなどは、比較的よくラベル分けとカテゴリが一致している（70~80%）。   
一方で、kaden-cghannel, livedoor-hommeなどはあまり一致していない。

比較的良い結果のカテゴリは、他のカテゴリとの違いがわかりやすい？　　   
たとえば映画系なら"映画", "公開"という単語は入ってそうだし、  
ガジェット系もスペック表示などがあるので、わかりやすいのかも。

topic-newsなどは時事ネタなので、精度が出づらいのは理解できるが、livedoor-homme （男性生活系）が、   
全然精度が出ない理由がわからない。Doc2Vecが適切に学習できていないのかもしれない。

ついでにPCAとt-SNEでもplotしてみた。
t-SNEは、上記の結果とconsistentなプロットになっているのがわかる。

### 結論
粗い定量評価なので、適切にモデルを評価できているとは言い難い。  
この評価方法は定量的というよりは、plotしつつ感覚をつかむ定性的な側面の意味合いが強い、と言える。

```python
ed.eval_km_cat()
ed.plot2d('PCA', n_sample=None, label_kind='category')
ed.plot2d('t-SNE', n_sample=3000, label_kind='category')
```

## 2. 教師ありモデル（Neural net）を用いた評価

- doc_vecをfeature、カテゴリを教師データとして、Neural netで学習する。
- doc2vecが適切に学習できていれば、正解率（accuracy）が高くなるはず。

### 結果＆結論
#### 2-1. 正解率による評価
正解率（accuracy）は79%  
悪くはない精度だが、間違いもまだまだあるという値。これは教師なしの評価の感覚ともconsistent。  
この評価方法は、doc2vecのモデル精度（= vector情報）を適切に評価できている、と言ってよさそう。

#### 2-2. 9つのカテゴリ毎にも評価値を算出
「smax, movie-enterは高く、livedoor-hommeは低い」など、  
評価方法1とconsistentな結果が得られている。

#### 2-3. 正しく予測した記事と、間違って予測した記事の比較
間違いのケースは２種類
1. doc2vecが適切に学習できていない。
2. 記事内容から、カテゴリ分けするのが難しいケース（教師カテゴリと記事内容が適切にマッチしていない, etc）

注意点としては、常に1のケースではないので（2のケースもあるので）、  
**doc2vecの評価を目的とする場合、正解率100%を目指すことは適切ではない。**    
むしろ100%近くなったらoverfit（記事内容ではなく、特定の単語やURLで"正解"を学習している）  
を疑った方がよさげ。  

**Parameter tuningするとき正解率を相対的に比較する**、などが適切な使い方といえる。

```python
# データ準備
import keras
from keras.models import Sequential
from keras.optimizers import Adam
from keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

test_ratio = 0.2

xs = ed.dvecs
le = preprocessing.LabelEncoder()
le.fit(ed.category_ary)
ys = le.transform(ed.category_ary)
num_classes = len(le.classes_)

x_train, x_test, y_train, y_test, indices_train, indices_test =\
    train_test_split(xs, ys, np.arange(ys.size), test_size=test_ratio, random_state=0, stratify=ys)
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
# モデルのアーキテクチャ構築
# 標準的なMLPを使用

l2_val = 0.01

model = Sequential()
model.add(Dense(30, activation='relu', input_shape=(xs[0].size,), kernel_regularizer=keras.regularizers.l2(l2_val)))
model.add(Dropout(0.2))
model.add(Dense(30, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))
```

```python
# トレーニング　＆　結果
batch_size = 128
epochs = 1000

model.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])
history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, verbose=0, validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=0)
```

```python
# 評価値を表示
from sklearn.metrics import classification_report
print('  Test loss:', score[0])
print('  Test accuracy:', score[1])
print('  Epoch:', history.epoch[-1])

y_test_cat = np.argmax(y_test, axis=1)
y_pred = model.predict_classes(x_test)
print('\n', classification_report(y_test_cat, y_pred, target_names=le.classes_))

fig, ax1 = plt.subplots()
c1 = 'tab:blue'; c2 = 'tab:red'
ax1.plot(history.epoch, history.history['val_acc'], color=c1)
ax1.set_xlabel('epoch'); ax1.set_ylabel('val_acc', color=c1)
ax2 = ax1.twinx()
ax2.plot(history.epoch, history.history['val_loss'], color=c2)
ax2.set_ylabel('val_loss', color=c2)
plt.show()
```

### 結論
正解しているケースでは、「そのカテゴリらしい」記事内容を適切に予測できている。  
ある程度の精度で予測できている。  
→doc2vecがある程度うまく学習できている、ことがわかる。

間違いのケースは２種類ある。  
1. doc2vecが適切に学習できていない。
2. 記事内容から、カテゴリ分けするのが難しいケース（教師カテゴリと記事内容が適切にマッチしていない, etc）

注意点としては、常に1のケースではないので（2のケースもあるので）、  
**doc2vecの評価を目的とする場合、正解率100%を目指すことは適切ではない。**    
むしろ100%近くなったらoverfit（記事内容ではなく、特定の単語やURLで"正解"を学習している）  
を疑った方がよさげ。  
たとえば、livedoor-hommeだと、記事に'livedoor HOMME'と書いてあるケースが多そう。  
この単語に着目すれば「教師あり学習」の精度は上げられる。しかし、ここでの**本来の評価目標は、  
「doc2vecが記事内容全体を適切に情報化できるているか」**なので、'livedoor HOMME'という単語に着目した  
学習は適切とは言えない。手段が目的化しないよう、評価するときには注意が必要だろう。

**Parameter tuningするとき正解率を相対的に比較する**、などが適切な使い方といえる。

```python
# 準備
from collections import defaultdict
pred_cats = model.predict_classes(x_test)
y_test_int = [list(y).index(1) for y in y_test]

dic_test_results = defaultdict(dict)
for pred_cat, real_cat, index in zip(pred_cats, y_test_int, indices_test):
    tagged_words = tagged_words_list[index]
    words = tagged_words[0]
    tag = tagged_words.tags[0]
    dic_test_results[tag]['words'] = words
    dic_test_results[tag]['pred_cat'] = le.inverse_transform(pred_cat)
    dic_test_results[tag]['real_cat'] = le.inverse_transform(real_cat)
    dic_test_results[tag]['correct'] = pred_cat == real_cat
```

```python
# 正解した記事と間違った記事を表示
n_show = 10
seed = 0

dic_correct = {k: v for k, v in dic_test_results.items() if v['correct']==1}
dic_wrong = {k: v for k, v in dic_test_results.items() if v['correct']==0}

random_wrong_tags = np.random.permutation(list(dic_wrong.keys()))

def print_results(dic, n_show, seed=0):
    np.random.seed(seed)
    random_tags = np.random.permutation(list(dic.keys()))
    for tag in random_tags[:n_show]:
        result = dic[tag]
        template = 'tag: {tag}\npred_cat: {pred_cat}\nreal_cat: {real_cat}\nwords:{words}'
        print(template.format(tag=tag, **result), end='\n\n')

#予測が正解した記事
print('\n=== Correct predictions ===')
print_results(dic_correct, n_show, seed)

#予測が間違った記事
print('\n=== Wrong predictions ===')
print_results(dic_wrong, n_show, seed)
```
