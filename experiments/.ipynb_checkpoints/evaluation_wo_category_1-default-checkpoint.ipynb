# 定性的な評価方法

## 目的
カテゴリデータを用いずに（= 教師なしデータの場合として扱い）評価を行う。

## モデル
doc2vec_pp1_setup1-default.model  
(train_doc2vec_1-default.ipynbでトレーニングしたモデル)

## 評価方法
#### 評価方法1
**「学習したベクトルのcosine類似度」と「文章の類似度」に相関があるか、人が見て判断**
- 「学習したベクトルの類似度」は計算可能。「文章の類似度」は、人が判断する。

#### 評価方法2
**学習したベクトルをKmeansでクラスタリング & 2次元プロット**
- Doc2Vecが適切に学習していれば、文章ベクトルが(ある程度)クラスタ化すると期待される。
- このとき（= Doc2Vecが上手く学習できていた場合）、
    - doc_vectorをKmeansでクラスタリングすれば、似たような記事がクラスタ化するはず。
    - 2次元プロットにおいても、ベクトルがクラスタ化しているのが可視化されるはず。

## 結果＆結論
いずれの方法でも、（完璧ではにものの）良い感じで評価できている（結果の詳細は、下記のcellを参照）。  
ただし「どのくらい良いか」の判断は定性的評価なので、難しい。  

たとえば、今回の実験では、ベクトルと文章の内容の類似度が完璧に一致しているわけではない  
ことは確認できるので、この一致度を上げるようにdoc2vecを  トレーニングできたら、  
より良いdoc2vecといって良いだろう。  

```python
import sys
import os.path as osp
sys.path.append(osp.pardir)
import pickle
from utils import train_docs
from utils import evaluation_docs
from gensim.models.doc2vec import Doc2Vec

fn_model = './model/doc2vec_pp1_setup1-default.model'
fn_tagged_words_list = './preproc_data/words_list_mecab_pp1.pickle'
fn_category_map = './preproc_data/category_map_pp1.pickle'
# fn_model = './model/sample_doc2vec_pp1_setup1-1.model'
# fn_tagged_words_list = './preproc_data/sample_words_list_mecab_pp1.pickle'
# fn_category_map = './preproc_data/sample_category_map_pp1.pickle'
n_clusters = 20
n_each_show = 5
```

## 1. ベクトルの類似度と文章の類似度の相関関係の定性的な評価
### 評価方法
#### 1-1. 文章ベクトル類似度が高い文章  
学習した文章ベクトルをいくつか選び（文章A）、そのベクトルと類似度が高いベクトルの文章(文章B)を複数取得。    
文章Aと文章Bを人が読んで、内容が似ているかを判断する。  
学習が適切にできていれば、「類似度の高いベクトル ~ 内容の近い文章」になることが期待される。

#### 1-2. ランダムに選んだ文章  
学習した文章ベクトルをいくつか選び（文章A）、ランダムに選んだ複数の文章（文章B）とのベクトル類似度を計算する。  
文章Aと文章Bを人が読んで、ベクトル距離と類似度に相関関係が適切にあるかを判断。  
学習が適切にできていれば、  
「ベクトル類似度が高ければ内容が似た文章になり、ベクトル類似度が低ければ、内容が似ていない文章になっている」   
はず。


### 結果
#### 1-1.
- (1番最初の)文章Aの記事は「浅田真央」の話題。
    - 内容はスポーツ系の記事と思われる。
    - その記事のベクトルと類似度が高いと評価されたベクトルを持つ５つの文章（文章B）は、どれもスポーツ系の記事になっている。
- (3番目の)文章Aの記事は「ドラクエX」の話題。
    - ５つの文章Bのうち、2つはドラクエXの話題。
    - 残りもデジタル系（孫さんの記事が微妙にずれている気もするが）。
- 残りも（多少外している記事もあるが）、だいたい期待通りに学習できていることがわかる。

#### 1-2.
- （1番最初の）文章Aの記事は「浅田真央」の話題。
    - 文書Bのうち、最も近い類似度=0.95を出しているのはスポーツ記事。
    - 残りの記事も、0.8~0.9とそれなりに類似度が高いが、スポーツ記事ではなく女性系の記事。
    - 文章Aは雑誌「女性自身」の記事で女性読者向けなので、女性系の記事と類似度が高くなるのかもしれない。
- （2番目の）文章Aの記事は「通り魔事件の映画化」の話題。
    - 文書Bのうち、最も近い類似度=0.92を出しているのは綾瀬はるかの記事。芸能系で近しいと判断されたようだ。
    - 残りの文章Bの類似度は0.6~0.7と低め。実際、内容はデジタル系の記事で、文章Aの内容とは全く異なる。
- （5番目の）文章Aの記事は「サッカー」の話題。
    - 文書Bのうち、最も近い類似度=0.93を出しているのはサッカーの話題。
    - 類似度=0.88と高い記事も野球の話題で、スポーツ系の記事。
    - 唯一おかしいのは「女性のナマ足」の記事で、なぜか0.91と類似度が高い。「足」つながり？
    - 残りの記事は類似度が低く、実際に内容もサッカーと異なるので、いい感じ。
    
### 結論
上記の結果から、（完璧ではないものの）いい感じで評価できている、と言える。

```python
model = Doc2Vec.load(fn_model)
tagged_words_list = pickle.load(open(fn_tagged_words_list, 'rb'))
dic_category_map = pickle.load(open(fn_category_map, 'rb'))
ed = evaluation_docs.evaluation_unlabeled_docs(model, tagged_words_list, n_clusters)
```

```python
tags = ed.get_random_tags(n_choice=5, random_state=0)
# 選ばれた文章から最も近いベクトルの文章を、５つ選び表示する。
# 表示は、(tag, cosine類似度, 文章の内容）の順
ed.print_similarities(tags, 'most', n_each_show)

# 選ばれた文章とは無関係に、ランダムに選んだ記事を表示する。
# 表示は、(tag, cosine類似度, 文章の内容）の順
ed.print_similarities(tags, 'random', n_each_show)
```

## 2. 学習したベクトルをKmeansでクラスタリング & 2次元プロット

Doc2Vecが適切に学習していれば、文章ベクトルが(ある程度)クラスタ化すると期待される。  
このとき（= Doc2Vecが上手く学習できていた場合）、  
- doc_vectorをKmeansでクラスタリングすれば、似たような記事がクラスタ化するはず。
- 2次元プロットにおいても、ベクトルがクラスタ化しているのが可視化されるはず。
    
### 結果
2次元plotには、PCAとt-SNEの２種類を用いた。PCAではクラスタリングをうまく可視化できなかったが、  
t-SNEでは（ある程度）クラスタリングが可視化できた。特に周縁部ではきれいにクラスタ化されている（中心部は微妙）。　  

またKmeans (n_cluster=20)で学習を行った。学習で20に分割されたを、色分けして２次元プロットに表示した。  
t-SNEでは、ベクトルのクラスタリングとKmeansのラベル分けがよく合っていることがわかる。  

もうひとつ下のcellで、Kmeansでラベル分けした20クラスタから、5つの記事を表示している。  
たとえば、
- t-SNEできれいにクラスタリングされている（周縁部の）label=13~16は、映画、デジタル系にきれいに分かれている。
- 逆に中心部のlabel=5~10は、独身女性、独身男性、ビジネス、などの内容で切り分けが難しいものが多いように見える。

### 結論
評価１と同じく、（完璧ではないものの）いい感じで評価できている、と言える。

```python
# PCAとt-SNEで２次元プロット
ed.plot2d('PCA', n_sample=None)
ed.plot2d('t-SNE', n_sample=3000)
```

```python
# Kmeans(n_cluster=20)で学習。
# Kmeansの20のクラスタ毎の記事内容を、５つずつ表示。
ed.print_each_cluster(n_each_show=5)
```
